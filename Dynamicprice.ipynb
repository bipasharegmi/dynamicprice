```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns

```

```python
df= pd.read_csv("Amazon Sale Report.csv")
```

```python
df.head()
df.tail()
df.info()
```

```python
df.columns
```

```python
df.duplicated()
```

```python
df.isna().sum()
```

```python
x=df["Amount"].mean()
```

```python
df.fillna({"Amount":x},inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
label=LabelEncoder()
```

```python
for col in df.select_dtypes(include="object").columns:
    df[col]=label.fit_transform(df[col])
```
